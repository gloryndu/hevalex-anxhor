# Housing Data (home value)
### Data type: ZHVI all homes(SF, condo/co-op) seasonally adjusted,     GEOGRAPHY (city)
##### ZHVI: Zillow Home Value Index â€” it represents a typical home value (not average or median) in a region.
##### All Homes (SFR(single family residential), Condo/Co-op): Includes all residential property types:
##### Time Series: Values tracked monthly over time
##### Smoothed: Short-term fluctuations are reduced to make trends clearer
##### Seasonally Adjusted: Adjusted to remove regular seasonal effects (e.g., summer sales spikes)

```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
main_df = pd.read_csv("City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")
main_df = pd.DataFrame(main_df)
main_df
```

### Task: selecting the city in maryland by parsing the data by states 

```python
md_states = main_df[main_df['State']=='MD']
md_states
```

```python
baltimore = md_states[md_states['RegionName']=='Baltimore']
baltimore
```

```python
import matplotlib.pyplot as plt
```

# need to be able to explain this 

```python
value_cols = baltimore.columns[8:]  
baltimore_melted = baltimore.melt(
    id_vars=['RegionName'], 
    value_vars=value_cols,
    var_name='Date', 
    value_name='HomeValue'
)


baltimore_melted['Date'] = pd.to_datetime(baltimore_melted['Date'])

plt.figure(figsize=(12, 6))
plt.plot(baltimore_melted['Date'], baltimore_melted['HomeValue'])
plt.title('Baltimore Home Values Over Time')
plt.xlabel('Date')
plt.ylabel('ZHVI Home Value ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
